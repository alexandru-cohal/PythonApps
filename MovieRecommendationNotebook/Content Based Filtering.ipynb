# Content Based Filtering

### Load the data

```
import pandas as pd
```

```
movies = pd.read_csv("movies_small.csv", sep=";")

movies["overview"] = movies["overview"].fillna("")
```

```
movies
```

### Create the TFIDF (Term Frequency â€“ Inverse Document Frequency) matrix

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english")
tfids_matrix = tfidf.fit_transform(movies["overview"])

pd.DataFrame(tfids_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

### Create the Similarity matrix

```
from sklearn.metrics.pairwise import linear_kernel

similarity_matrix = linear_kernel(tfids_matrix, tfids_matrix)

similarity_matrix
```

### Find the most similar movies to a certain movie

```
movie_title = "John Carter"
idx_movie = movies.loc[movies["title"] == movie_title].index[0]
```

```
scores_movie = list(enumerate(similarity_matrix[idx_movie]))
scores_movie = sorted(scores_movie, key=lambda x: x[1], reverse=True)
scores_movie
```

```
idx_movies_similar = [tpl[0] for tpl in scores_movie[1:]]
idx_movies_similar
```

```
list(movies["title"].iloc[idx_movies_similar])
```

```
def similar_movies(movie_title, nr_movies):
    idx_movie = movies.loc[movies["title"] == movie_title].index[0]
    scores_movie = list(enumerate(similarity_matrix[idx_movie]))
    scores_movie = sorted(scores_movie, key=lambda x: x[1], reverse=True)
    idx_movies_similar = [tpl[0] for tpl in scores_movie[1:nr_movies + 1]]
    titles_movies_similar = list(movies["title"].iloc[idx_movies_similar])
    return titles_movies_similar
```

```
similar_movies("Kung Fu Panda 3", 3)
```
